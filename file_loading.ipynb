```julia
using BenchmarkTools
```

# julia file loading speed

## benchmarks

```julia
@benchmark begin
    open("Small.pgn") do f
        i = 1
        for k in readlines(f)
            i+=1
        end
    end
end
```

```julia
@benchmark begin
    open("Small.pgn") do f
        i = 1
        for k in eachline(f)
            println(k)
            i+=1
        end
    end
end
```

this produces chaotic results, and at random occasions has half the allocations of previous examples, but doesn't run any faster for some mysterious reason.

```julia
@benchmark begin
    open("Small.pgn") do f
        i = 1
        contents = IOBuffer(read(f, String))
        for k in readlines(contents)
            i+=1
        end
    end
end
```

```julia
@benchmark begin
    open("Small.pgn") do f
        i = 1
        contents = IOBuffer(read(f, String))
        for k in eachline(contents)
            i+=1
        end
    end
end
```

```julia
@benchmark begin
    open("Small.pgn") do f
        i = 1
        contents = IOBuffer(read(f, String))
        for k in eachline(contents)
            i+=1
        end
    end
end
```

```julia
@benchmark begin
    open("Small.pgn") do f
        i = 1
        contents = IOBuffer(read(f, String))
        for k in contents |> Base.Fix2(read, String)
            i+=1
        end
    end
end
```

```julia
global i = 1
@benchmark begin
    open("Small.pgn") do f
        global i = 0
        for k in read(f, String)
            global i += 1
        end
    end
end
```

```julia
global i = 1
@benchmark begin
    open("Small.pgn") do f
        for k in read(f, Char)
            # println(k)
            if k == "\n"
                global i += 1
            end
        end
    end
end
```

## Conclusion

Both `readlines` and `eachline` are a massive bottleneck(I presume that's because they allocate for each line).
`IOStream` created from a `String` didn't even complete, it suggests to me I've used it incorrectly,
I recommend using `read(open(filename, "r"), String)` for the absolute fastest solution, but `read(::IOBuffer, String)` can be used too.
